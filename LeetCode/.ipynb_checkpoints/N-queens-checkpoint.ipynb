51. N-Queens<br>

The n-queens puzzle is the problem of placing n queens on an n x n chessboard such that no two queens attack each other.<br>

Given an integer n, return all distinct solutions to the n-queens puzzle. You may return the answer in any order.<br>

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' both indicate a queen and an empty space, respectively.<br>

Resource:
https://leetcode.com/problems/n-queens/

```python
class Solution(object):
    def solveNQueens(self, n):
        """
        :type n: int
        :rtype: List[List[str]]
        """
        result = []
        combination = []
        self.backtrack(result, combination, n)
        return result
    def backtrack(self, result, combination, n):
        cur_row = len(combination)
        if cur_row == n:
            result.append(self.draw(combination[:], n))
            return
        for cur_col in range(n):
            if self.check(cur_row, cur_col, combination):
                combination.append(cur_col)
                self.backtrack(result, combination, n)
                combination.pop()
    def check(self, cur_row, cur_col, combination):
        for row, col in enumerate(combination):
            if cur_col == col:
                return False
            if row + col == cur_row + cur_col or row - col == cur_row - cur_col:
                return False
        return True
    def draw(self, combination, n):
        board = []
        for row in range(n):
            cur_row = []
            for col in range(n):
                if combination[row] == col:
                    # Q
                    cur_row.append('Q')
                else:
                    cur_row.append('.')
            # ['.', 'Q', '.', '.'] -> '.Q..'
            cur_row_str = ''.join(cur_row)
            board.append(cur_row_str)
        return board
```
