```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
'let us assume we have 3 clusters'
k = 3
```

```python
def init(k):
    'init centers'
    return X[np.random.choice(len(X), k)]
```

```python
'generate k clusters'
c = [[0, 0], [10, 10], [5, 5]]
X, y = make_blobs(n_samples = 50, centers = c, n_features = 2, random_state = 0)
```

```python
centers = init(k)
```

```python
def plot(X, centers):
    'we will use this function to keep updating the graph'
    plt.scatter(X[:,0], X[:,1], c = y)
    plt.scatter(centers[:,0], centers[:, 1], c = 'r')
    plt.annotate('centers', xy = (centers[0, 0], centers[0, 1]), 
                 arrowprops = dict(facecolor='black', shrink=0.05), xytext = (centers[0, 0] + 2, centers[0, 1] + 4))
    plt.show()
```

```python
plot(X, centers)
```

## One iteration

```python
def cluster(X, centers):
    '''return the cluster each point belong to
    if you dont understand this function
    please visit https://github.com/d5423197/LinkedIn/blob/main/KMeans_implementation/Numpy%20broadcasting%20example.ipynb
    '''
    return np.argmin(np.sqrt(((X[:, np.newaxis] - centers)**2).sum(axis = 2)), axis = 1)
```

```python
label = cluster(X, centers)
label
```

```python
def recenter(label, X, centers, k):
    'this function is used to recalculate the centers'
    result = []
    'label starts from 0 to k-1'
    for i in range(k):
        'points belong to this cluster'
        points = X[label == i]
        new_center = points.mean(axis = 0)
        result.append(new_center)
    return np.stack(result, axis = 0)
```

```python
'new centers'
centers = recenter(label, X, centers, 3)
centers
```

```python
plot(X, centers)
```

## Run many iterations

```python
for i in range(10000):
    label = cluster(X, centers)
    centers = recenter(label, X, centers, 3)
```

```python
plot(X, centers)
```
